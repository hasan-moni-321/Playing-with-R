```python
library(datasets)
library(ggplot2)
```

```python
titanic <- read.csv("/home/hasan/Data Set/titanic/train.csv")
head(titanic)
```

```python
# setup factor
titanic$Pclass <- as.factor(titanic$Pclass)
titanic$Survived <- as.factor(titanic$Survived)
titanic$Sex <- as.factor(titanic$Sex)
titanic$Embarked <- as.factor(titanic$Embarked)
```

```python
ggplot(titanic, aes(x=Survived)) + geom_bar()
```

```python
# percentage of survived columns 
percent_survived <- prop.table(table(titanic$Survived))
percent_survived 
```

```python
# add some customization for labels and themes
ggplot(titanic, aes(x=Survived)) + 
        theme_bw() +
        geom_bar() + 
        labs(y='passenger count',
             x='titanic survived')
```

```python
# Survived rate by gender
ggplot(titanic, aes(x=Sex, fill=Survived)) +
        theme_bw() +
        geom_bar() +
        labs(y='passsenger count',
            title='titanic survive rate by sex')
```

```python
# survived rate by class of ticket
ggplot(titanic, aes(x=Pclass, fill=Survived)) +
        theme_bw() +
        geom_bar() +
        labs(y='passenger count',
            title='titanic survived rate by pclass')
```

```python
# Survived rate by class of ticket and gender
ggplot(titanic, aes(x=Sex, fill=Survived)) +
        theme_bw() +
        facet_wrap(~ Pclass) + 
        geom_bar() + 
        labs(y='passenger count',
             title='titanic survived rate by Pclass and Sex')
```

```python
# Distribution of the passenger ages
ggplot(titanic, aes(x=Age)) +
theme_bw() +
geom_histogram(binwidth=5) +
labs(y='passenger count',
     x='age (binwidth=5)',
     title='titanic age distribution')
```

```python
# survived rate by age
ggplot(titanic, aes(x=Age, fill=Survived)) +
theme_bw() +
geom_histogram(binwidth=5) +
labs(y='passenger count',
     x='age (binwidth=5)',
     titanic='survived rate by age')
```

```python
# box and whisker plot
ggplot(titanic, aes(x=Survived, y=Age)) +
        theme_bw() +
        geom_boxplot() +
        labs(y='Age',
             x='Survived',
             title='Titanic survived rate by age')
```

```python
# survived rate by age when segmented by gender and class of ticket
ggplot(titanic, aes(x=Age, fill=Survived)) + 
        theme_bw() +
        facet_wrap(Sex ~ Pclass) +
        geom_density(alpha=0.5) + 
        labs(y='Density', 
             x='Age',
             title='Titanic survived rate by age, pclass and sex')
```

```python
# histogram
ggplot(titanic, aes(x=Age, fill=Survived)) +
        theme_bw() +
        facet_wrap(Sex ~ Pclass) +
        geom_histogram(binwidth=5) +
        labs(y='count',
             x='Age',
             title='titanic survived rate by age pclass and sex')
```
